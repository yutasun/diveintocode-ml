# 【問題1】データの取得
irisデータセット読み込み,説明変数を下のようなPandasのDataFrame型変数Xに格納

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris_data = load_iris()
X = iris_data.data
y = iris_data.target
data_names = ['sepal_length','sepal_width','petal_length','petal_width']
target_names = ['setosa', 'versicolor', 'virginica']
#print(type(X))
# print(type(target_names))

df_X = pd.DataFrame(X,columns = data_names)
df_y = pd.DataFrame(y,columns = ['Species'])
print(df_X)
print(df_y)
```

# 【問題2】データの結合
X,yを結合させたDataFrame型変数dfを作成

```python
df = pd.concat([df_X, df_y], axis=1,)
print(df)
```

# 【問題7】特徴量間の関係の確認
<br>縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
<br>全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
<br>４つの特徴量に関する相関係数行列を作る
<br>相関係数行列のヒートマップを作る

```python
def scatter_make(x_line,y_line): #縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
    x = x_line
    y = y_line
    plt.scatter(x, y)# 散布図を描画

scatter_make(feature_sepal_length,feature_sepal_width)
scatter_make(feature_sepal_length,feature_petal_length)
scatter_make(feature_sepal_length,feature_petal_width)
scatter_make(feature_sepal_width,feature_petal_length)
scatter_make(feature_sepal_width,feature_petal_width)
scatter_make(feature_petal_length,feature_petal_width)
```

```python
import matplotlib.pyplot as plt #全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
import seaborn as sns

df_soukan = sns.load_dataset("iris")
sns.pairplot(df_soukan, hue="species", size=2.5)
plt.show()
```

```python
df_soukan.corr(method='pearson') #４つの特徴量に関する相関係数行列を作る 
```

```python
sns.heatmap(df.corr(method='pearson'),
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=df.corr(method='pearson').columns.values,
            yticklabels=df.corr(method='pearson').columns.values
           )
plt.show()
```
