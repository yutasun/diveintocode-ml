【問題１】　富士山を超える最小の折る回数

```python
THICKNESS = 0.00008
sum_list=[THICKNESS] #確認用のリスト宣言

#関数定義　富士山（引き数：hight）を超える折る回数（返り値：folded_times）
def proceed_mount(hight):
    sum = THICKNESS
    folded_times = 0
    
    while(sum < hight):
        sum = sum * 2
        folded_times += 1
        sum_list.append(sum) #確認用のリスト生成
    print(sum_list) #確認用のリスト表示
    return folded_times

print("{}回".format(proceed_mount(3778)))
```

```python
#関数の定義
def proceed_mount(thickness,distance):
    sum = thickness
    folded_times = 0
    sum_list=[thickness] #確認用のリスト宣言

    while(sum < distance):
        sum = sum * 2
        folded_times += 1
        sum_list.append(sum) #確認用のリスト生成
    print(sum_list) #確認のためのリスト表示
    return folded_times

print("{}回".format(proceed_mount(0.00008,4.243*9460730472580800)))
```

【問題３】必要な紙の長さ

```python
import math

#折る回数を求める関数（問題２と同じ関数）
def proceed_mount(thickness,distance):
    sum = thickness
    folded_times = 0
    sum_list=[thickness] #確認用のリスト宣言

    while(sum < distance):
        sum = sum * 2
        folded_times += 1
        sum_list.append(sum) #確認用のリスト生成
    print(sum_list) #確認のためのリスト表示
    return folded_times

#必要な長さを求める関数
def required_length(thickness,folded_times): 
    length = math.pi*thickness/6*(2**folded_times+4)*(2**folded_times+1)
    return length

# print("{}回".format(proceed_mount(0.00008,3778)))
print("★必要な紙の長さ・富士山の場合：{:2e}km★".format(required_length(0.00008,proceed_mount(0.00008,3778))))
print("★必要な紙の長さ・月の場合:{:.2e}km★".format(required_length(0.00008,proceed_mount(0.00008,38440000))))
print("★必要な紙の長さ・恒星の場合:{:2e}km★".format(required_length(0.00008,proceed_mount(0.00008,4.243*9460730472580800))))
```
