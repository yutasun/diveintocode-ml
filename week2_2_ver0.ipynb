# 【問題4】山を下る関数を作成

```python
def des_pos_FUJI(pos):
    m_list = [pos]
    a = int(nextpos_FUJI(pos))
    m_list.append(a)
    for i in range(1000):
        if np.round(-0.2*gradientFUJI(m_list[-1])) != 0:
            a = int(nextpos_FUJI(m_list[-1]))
            m_list.append(a)
    return a,m_list

des_pos_FUJI(136)
```
