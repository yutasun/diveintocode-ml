# 【問題3】Iris（2値分類）をKerasで学習

```python
# 入力データ（２値分類）
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
# データセットの読み込み

iris_data = load_iris()
X = iris_data.data[50:]
y = iris_data.target[50:] - 1

y = y.astype(np.int)[:, np.newaxis]

# trainとtestに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# さらにtrainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
model = tf.keras.Sequential([
            tf.keras.layers.Dense(50, activation = tf.nn.relu, input_shape=(4,)),
            tf.keras.layers.Dense(100, activation = tf.nn.relu, input_shape=(50,)),
            tf.keras.layers.Dense(1, activation = tf.nn.sigmoid)])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=tf.train.AdamOptimizer(learning_rate=0.01),
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                    batch_size=10,
                    epochs=10,
                    verbose=1)
```

```python
y_pred_proba = model.predict(X_train)[:, 0]
# 確率を0, 1に変換
y_pred = np.where(y_pred_proba >0.5, 1, 0)
print("y_pred_proba", y_pred_proba)
print("y_pred", y_pred)
```

```python
from sklearn import metrics
print(str(metrics.accuracy_score(y_train, y_pred)))
```

# 【問題4】Iris（多値分類）をKerasで学習

```python
iris_data = load_iris()
X = iris_data.data[:]
y = iris_data.target[:]
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
y_one_hot = enc.fit_transform(y[:, np.newaxis])
y = y_one_hot.astype(np.int)
# print(y.shape)
# trainとtestに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# さらにtrainとvalに分割
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
model2 = tf.keras.Sequential([
            tf.keras.layers.Dense(50, activation = tf.nn.relu, input_shape=(4,)),
            tf.keras.layers.Dense(100, activation = tf.nn.relu, input_shape=(50,)),
            tf.keras.layers.Dense(3, activation = tf.nn.softmax)])
```

```python
model2.compile(loss='categorical_crossentropy',
              optimizer=tf.train.AdamOptimizer(learning_rate=0.01),
              metrics=['accuracy'])
```

```python
history = model2.fit(X_train, y_train,
                    batch_size=10,
                    epochs=10,
                    verbose=1)
```

```python
y_pred_proba = model2.predict(X_train)
# 確率を0, 1に変換
y_pred = np.argmax(y_pred_proba,1)
print("y_pred", y_pred)
```

```python
np.argmax(y_train, axis = 1)
```

```python
print(str(metrics.accuracy_score(np.argmax(y_train, axis = 1), y_pred)))
```

# 【問題6】MNISTをKerasで学習

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train.astype(np.float)
X_test = X_test.astype(np.float)
X_train /= 255
X_test /= 255
print(X_train.max()) # 1.0
print(X_train.min()) # 0.0
enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
y_train_one_hot = enc.fit_transform(y_train[:, np.newaxis])
y_test_one_hot = enc.transform(y_test[:, np.newaxis])
print(y_train.shape) # (60000,)
print(y_train_one_hot.shape) # (60000, 10)
print(y_train_one_hot.dtype) # float64
```

```python
X_train_0, X_val_0, y_train_0, y_val_0 = train_test_split(X_train, y_train_one_hot, test_size=0.95)

X_train_0 = X_train_0[:, :, :, np.newaxis]
X_val_0 = X_val_0[:, :, :, np.newaxis]

print(X_train_0.shape) # (600,1, 28,28)
print(X_val_0.shape) # (59400,1, 28,28)
print(y_train_0.shape) # (600, 10)
print(y_val_0.shape) # (59400, 10)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, Activation

model4 = Sequential()
model4.add(Conv2D(16, (3, 3), padding='same',activation='relu', input_shape=(28,28,1)))
                           # input_shape は指定しなくてもよいが，summary()を使うためには必要
# model4.add(Conv2D(16, (3, 3), padding='same',activation='relu'))
model4.add(Conv2D(16, (2, 2), strides=2, padding='same'))


model4.add(Conv2D(32, (3, 3), padding='same',activation='relu'))
# model4.add(Conv2D(32, (3, 3), padding='same',activation='relu'))
model4.add(Conv2D(32, (2, 2), strides=2, padding='same'))

model4.add(Conv2D(64, (3, 3), padding='same',activation='relu'))
# model4.add(Conv2D(64, (3, 3), padding='same',activation='relu'))
model4.add(Conv2D(64, (2, 2), strides=2, padding='same'))

model4.add(Flatten())
# model4.add(Dense(512, activation='relu'))
model4.add(Dense(10))
model4.add(Activation('softmax'))
 
# コンパイル
model4.compile(loss='categorical_crossentropy',optimizer='SGD',metrics=['accuracy'])
```

```python
model4.summary()
```

```python
history = model4.fit(X_train_0, y_train_0,
                    batch_size=20,
                    epochs=10,
                    verbose=1)
```

```python
y_pred = model4.predict(X_val_0)
print("y_pred", y_pred)
```

```python
print(str(metrics.accuracy_score(np.argmax(y_val_0, axis = 1), np.argmax(y_pred, axis = 1))))
```
